# Iris Dataset Overview

The Iris dataset is a well-known dataset in the field of machine learning and statistics. It contains 150 instances, each representing an individual flower sample from one of three different species of the Iris flower.

## Class Distribution
- **Iris Setosa**: 50 instances
- **Iris Versicolour**: 50 instances
- **Iris Virginica**: 50 instances

## Attribute Information
The dataset features four numerical attributes and one categorical attribute for the class label, as follows:

1. **Sepal Length (cm)**: The length of the sepal, which is the outer part of the flower.
2. **Sepal Width (cm)**: The width of the sepal.
3. **Petal Length (cm)**: The length of the petal, which is the inner part of the flower.
4. **Petal Width (cm)**: The width of the petal.

## Class Labels
The dataset categorizes the flower samples into three species:

- **Iris Setosa**: Small and distinctively shaped petals and sepals.
- **Iris Versicolour**: Intermediate dimensions, not linearly separable from Iris Virginica.
- **Iris Virginica**: Large petals and sepals, distinguishable from Iris Setosa but not easily separable from Iris Versicolour.

# Loading the dataset

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
# delete a column
df = df.drop(columns = ['Id'])
df.head()
```

# Exploratory Data Analysis

```python
# Create a pairplot to visualize distributions and relationships between features
sns.pairplot(df, hue='Species', markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# Create box plots to compare feature distributions across species
for column in df.columns[:-1]:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Species', y=column, data=df)
    plt.title(f'Box plot of {column} by Species')
    plt.show()
```

```python
# Compute the correlation matrix for numeric columns only
corr_matrix = df.drop(columns='Species').corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Iris Dataset')
plt.show()

```
